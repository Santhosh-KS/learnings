```python
def plotYear(year):
    data = gapminder[gapminder.year == year]
    area = 5e-6 * gapminder.population
    colour = gapminder.region.map({'Africa': 'skyblue', 'America':'palegreen', 'Europe':'gold', 'Asia':'coral'})
    data.plot.scatter('babies_per_woman', 'age5_surviving', s=area, c=colour, linewidth=1, edgecolor='k',
                     figsize=(12,9))
    
    pp.axis(ymin=50,ymax=105, xmin=0,xmax=8)
    pp.xlabel('Babies per Woman')
    pp.ylabel('% of Babies surviving at Age 5')
```

```python
plotYear(1965)
```

```python
interact(plotYear, year = widgets.IntSlider(min=1950, max=2015, step=1, value=1965))
```
